```python
import pandas as pd

pts = ['Sarcoma167GEX', 'Sarcoma322GEX', 'Sarcoma559GEX',
              'Sarcoma708GEX', 'SarcomaS408GEX', 'SarcomaS410GEX', 
              'SarcomaS914GEX', 'SarcomaS956GEX']
```

```python
pts
```

```python
arrs = []
for x in pts:
    arrs.append(pd.read_csv('DiffusionAnalysisData/' + x + 'metadata.csv').set_index("Unnamed: 0")) 
```

```python
arrs[0]
```

```python
all_non_tumor = pd.read_csv("DiffusionAnalysisData/all_non_tumor_BI_general_assignments_meta_data.csv").set_index("Unnamed: 0")
```

```python
from collections import Counter
Counter(all_non_tumor['BI_broad_assignments_v2'])
```

```python
all_non_tumor["new_ind"] = all_non_tumor.apply(lambda x: x.name.split("_")[0], axis =1)
```

```python
def add_label(row):
    temp = all_non_tumor[all_non_tumor['new_ind'] == row.name]
    temp = temp[temp['orig.ident'] == row['orig.ident']]
    if (len(temp) == 0):
        return ("Tumor cell")
    else:

        return(list(temp['BI_broad_assignments_v2'])[0])
```

```python
for i,x in enumerate(pts):
    print(i)
    arrs[i]["bar_assignments"] = arrs[i].apply(lambda x: add_label(x), axis = 1)
```

```python
cols = []
index = []
for x in arrs:
    cols.append(x['bar_assignments'].value_counts(normalize=True))
    index.append(list(x["orig.ident"])[0])
plot = pd.DataFrame(cols).fillna(0)
plot.index = index
plot
```

```python
plot.index


plot = plot.reindex(index = ['Sarcoma167GEX', 'SarcomaS410GEX', 'SarcomaS914GEX', 
                             'Sarcoma322GEX', 'Sarcoma559GEX', 'Sarcoma708GEX',
       'SarcomaS408GEX',  'SarcomaS956GEX'
                            ])
plot


plot.index = plot.index.str.replace('Sarcoma', 'S')
plot.index = plot.index.str.replace('SS', 'S')


plot
color = ['#f6766d', '#c49a00', '#53b400', '#3bc193', '#40b6eb','#a58aff','#f761d7']
colnames = list(plot.columns)
color_dict = dict(zip(colnames, color))
color_dict
plot
new_order = ['Tumor cell', 'T-cell', 'Endothelial cell', 'Myeloid',
             'Fibroblast', 'B-cell/Plasma cell', 'Myofibroblast']
plot
```

```python
plot = plot[new_order]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```
