## Annotate MT, RPS, and RPL genes

Mitochondrial (MT) and Ribosomal (RPS & RPL) gene names begin with specific prefixes. We use this to calculate the percentage of these types of genes (features) in each cell (identity).

```python
%%R
for(i in 1:length(seu_raw.list)) {
    seu_raw.list[[i]][['percent.mt']] <- PercentageFeatureSet(seu_raw.list[[i]], pattern = '^MT-')
    seu_raw.list[[i]] <- PercentageFeatureSet(seu_raw.list[[i]], pattern = '^RPS', col.name = 'percent.rps')
    seu_raw.list[[i]] <- PercentageFeatureSet(seu_raw.list[[i]], pattern = '^RPL', col.name = 'percent.rpl')
    seu_raw.list[[i]]$percent.rp <- seu_raw.list[[i]]$percent.rps + seu_raw.list[[i]]$percent.rpl
}
```

## Identify doublets using scrublet

```python
import scrublet as scr
import scipy.io
import sys
```

```python
# read counts matrix file written by writeMM above

counts_list = []
for pat in pat_list:
    counts_matrix = scipy.io.mmread(''.join(['data/', pat, '/matrix_', pat, '_raw.mtx'])).T.tocsc()

    # run scrublet using expected doublet rate
    scrub = scr.Scrublet(counts_matrix, expected_doublet_rate=doublet_rate)
    doublet_scores, predicted_doublets = scrub.scrub_doublets(min_counts=1, min_cells=1,
                                                              min_gene_variability_pctl=50,
                                                              n_prin_comps=20)
    print(doublet_scores, predicted_doublets)
    
    
    with open(''.join(['data/', pat, '/doublets_', pat, '_raw.txt']),'w') as outFI:
        outFI.write('predicted_doublets\tdoublet_scores\n')
        for j in range(len(predicted_doublets)):
            outFI.write(str(predicted_doublets[j]).upper()+'\t'+str(doublet_scores[j])+'\n')
```

## Run Seurat workflow excluding doublets

Read doublet data from file

```python
%%R

seu_filtered.list <- lapply(X = seu_raw.list, FUN = function(x) {
    pat <- Project(x)
    doublets <- read.table(paste0('data/', pat, '/doublets_', pat, '_raw.txt'), header = T)
    x[['predicted_doublets']] <- doublets$predicted_doublets
    x[['doublet_scores']] <- doublets$doublet_scores

    # remove intermediate files
#     system(paste0('rm data/', pat, '/matrix_', pat, '_raw.mtx'))
#     system(paste0('rm data/', pat, '/doublets_', pat, '_raw.txt'))

    # Subset:
    #   1. unique genes: between 200 and 7500
    #   2. gene counts: between 400 and 40000
    #   3. mitochondrial genes: < 10%
    #   4. no doublets
    minFeature <- 300
    maxFeature <- 7500
    minCount <- 600
    maxCount <- 40000
    maxMT <- 10
    x <- subset(x, subset = nFeature_RNA > minFeature & nFeature_RNA < maxFeature & nCount_RNA > minCount &
                nCount_RNA < maxCount & percent.mt < maxMT & predicted_doublets == F)
})
```

```python
%%R

# Seurat workflow
seu.list <- lapply(X = seu_filtered.list, FUN = function(x) {
    pat <- Project(x)

    # Normalize counts using default LogNormalize method with a scale factor of 10000
    x <- NormalizeData(x)

    # Identify outlier features
    x <- FindVariableFeatures(x)
        
    # Center expression of each gene, so the mean across all cells is 0
    # Scale the expression of each gene, so that the variance across cells is 1
    # Results are in seu[["RNA"]]@scale.data
    x <- ScaleData(x)

    # Run PCA on 50 dimensions (default)
    x <- RunPCA(x)
    x <- JackStraw(x, num.replicate = 100)
    x <- ScoreJackStraw(x, reduction = 'pca', dims = 1:20)

    # Compute k-nearest neighbors for k=20 (default) for first 30 dims
    x <- FindNeighbors(x, dims = 1:30)

    # Cluster using Shared Nearest Neighbor (SNN), updates 'seurat_clusters'
    x <- FindClusters(x)

    # Run UMAP dimensional reduction
    x <- RunUMAP(x, dims = 1:30)
})
```
