```python
scores = []
for x in MPs:
    temp = {}
    for y in x:
        temp[y] = merged_dict[y]
    scores.append(temp)
```

```python
sample_scores = []
for data in scores:
    result = {}

    for key, value in data.items():
        prefix = key.split(".")[0]  # Extract the first 12 characters of the key
        if prefix in result:
            result[prefix] += value  # Add the value to the existing key
        else:
            result[prefix] = value  # Create a new key-value pair
    sample_scores.append(result)
```
